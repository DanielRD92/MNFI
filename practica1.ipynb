```python
import numpy as np
from math import pi
import matplotlib.pyplot as plt
def congruencial(x0,n):
    a=16807  
    b=0
    m=2**31-1
    x=[]
    for i in range(0,n):
        x1=x0*a+b
        x0=x1%m
        x.append(x0/m)
    return (x0,x)

def boxmuller(x0,n):
    y=[]
    x=[]
    u1=np.array(congruencial(x0,n)[1])
    u2=np.array(congruencial(congruencial(x0,n)[0],n)[1])
    x=np.sqrt(-2*np.log(u1))*np.cos(2*pi*u2)
    y=np.sqrt(-2*np.log(u1))*np.sin(2*pi*u2)
    return (x,y)

def boxmullerantiteticas(x0,n):
    y=[]
    x=[]
    u1=np.array(congruencial(x0,n)[1])
    u2=np.array(congruencial(congruencial(x0,n)[0],n)[1])
    x=np.sqrt(-2*np.log(1-u1))*np.cos(2*pi*(1-u2))
    y=np.sqrt(-2*np.log(1-u1))*np.sin(2*pi*(1-u2))
    return (x,y)
```

```python
x0=123456789
n=1000
x = boxmuller(x0,n)[0]
y = boxmuller(x0,n)[1]
xanti = boxmullerantiteticas(x0,n)[0]
yanti = boxmullerantiteticas(x0,n)[1]
[np.mean(x),len(x)*np.var(x)/(len(x)-1),np.mean(y),len(y)*np.var(y)/(len(y)-1)]
len(x)
```
